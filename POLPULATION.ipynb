```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the Excel file
df = pd.read_csv("/content/WPP2022_PopulationExposureBySingleAgeSex_Medium_1950-2021.csv")

# Convert columns to numeric data types and handle missing values (replace with median age)
df['SortOrder'] = pd.to_numeric(df['SortOrder'], errors='coerce')
df['LocID']= pd.to_numeric(df['LocID'],errors='coerce')
df['ISO3_code'] = pd.to_numeric(df['ISO3_code'], errors='coerce')

# Replace missing age values with the median age of the population
median_age_column5 = df['SortOrder'].median()
median_age_column6 = df['LocID'].median()
median_age_column7 = df['ISO3_code'].median()

df['SortOrder'].fillna(median_age_column5, inplace=True)
df['LocID'].fillna(median_age_column6, inplace=True)
df['ISO3_code'].fillna(median_age_column7, inplace=True)

# Define age categories
age_categories = ['0-14', '15-64', '65+']
# Create a new DataFrame with columns for each age category
df_stacked = df[['SortOrder', 'LocID', 'ISO3_code']].copy()
df_stacked.columns = age_categories

# Calculate the sum of each age category
total_population = df_stacked.sum()

# Create a bar chart for age distribution
plt.figure(figsize=(10, 6))
plt.bar(age_categories, total_population)
plt.xlabel('Age Category')
plt.ylabel('Population Count')
plt.title('Bar Chart for Age Distribution in the Population (2022)')
plt.show()
```
